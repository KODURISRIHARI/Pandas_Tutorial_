Name - The name of the Pokemon.
Type 1	The primary type of the Pokemon (e.g., Fire, Water, Grass).

Type 2 - The secondary type of the Pokemon (e.g., Flying, Poison, Psychic). Some Pokemon only have one type.

HP -	The Hit Points of the Pokemon, representing its health.

Attack -	The base attack stat of the Pokemon, influencing its physical damage output.

Defense -	The base defense stat of the Pokemon, influencing its resistance to physical attacks.

Sp. Atk	 - The Special Attack stat of the Pokemon, influencing its special damage output.

Sp. Def	 - The Special Defense stat of the Pokemon, influencing its resistance to special attacks.

Speed	 - The speed stat of the Pokemon, determining which Pokemon attacks first in battle.

Generation -	The generation the Pokemon was introduced in (e.g., Generation 1, Generation 2).

Legendary - A boolean value indicating whether the Pokemon is legendary (True) or not (False).
