```python
import random

# Predefined lists of names, addresses, and designations
names = [
    "Liam Smith", "Noah Johnson", "Oliver Brown", "Elijah Davis", "James Wilson",
    "William Garcia", "Benjamin Martinez", "Lucas Anderson", "Henry Taylor", "Alexander Thomas"
]

addresses = [
    "123 Main St, Springfield", "456 Elm St, Metropolis", "789 Oak St, Gotham",
    "321 Pine St, Star City", "654 Maple St, Central City",
    "987 Birch St, Smallville", "135 Cedar St, Coast City", "246 Walnut St, Happyville",
    "369 Cherry St, Hill Valley", "159 Ash St, Riverdale"
]

genders = ['Male', 'Female', 'Other']
designations = ['Software Engineer', 'Data Scientist', 'Manager', 'Analyst', 'Intern']

# Function to generate a single record
def generate_record():
    name = random.choice(names)
    gender = random.choice(genders)
    email = f"{name.replace(' ', '').lower()}@example.com"  # Simple email generation
    address = random.choice(addresses)
    designation = random.choice(designations)
    salary = random.randint(30000, 120000)  # Salary between 30,000 and 120,000
    age = random.randint(22, 60)  # Age between 22 and 60
    mobile = f"+91-{random.randint(1000, 9999)}-{random.randint(1000, 9999)}"  # Mobile number format

    return {
        'Name': name,
        'Gender': gender,
        'Email': email,
        'Address': address,
        'Designation': designation,
        'Salary': salary,
        'Age': age,
        'Mobile': mobile
    }

# Generate a list of 40 records
dummy_dataset = [generate_record() for _ in range(40)]

# Print the dummy dataset
df = pd.DataFrame(dummy_dataset)
df
```

```python
import random
from faker import Faker

# Initialize the Faker library
fake = Faker()

# Define possible values for Gender and Designation
genders = ['Male', 'Female', 'Other']
designations = ['Software Engineer', 'Data Scientist', 'Manager', 'Analyst', 'Intern']

# Function to generate a single record
def generate_record():
    name = fake.name()
    gender = random.choice(genders)
    email = fake.email()
    address = fake.address().replace("\n", ", ")  # Replace newline with comma for better formatting
    designation = random.choice(designations)
    salary = random.randint(30000, 120000)  # Salary between 30,000 and 120,000
    age = random.randint(22, 60)  # Age between 22 and 60
    mobile = f"+91-{random.randint(10000, 99999)}-{random.randint(10000, 99999)}"  # Mobile number format

    return {
        'Name': name,
        'Gender': gender,
        'Email': email,
        'Address': address,
        'Designation': designation,
        'Salary': salary,
        'Age': age,
        'Mobile': mobile
    }

# Generate a list of 40 records
dummy_dataset = [generate_record() for _ in range(100)]

# Print the dummy dataset
df = pd.DataFrame(dummy_dataset)
df
```
